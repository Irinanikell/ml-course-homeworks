```python
import sys
import numpy as np
import pandas
import sklearn

print("1. python version: ", sys.version)
print("2. numpy version: ", np.__version__)
print("3. pandas version: ", pandas.__version__)
print("4. sklearn version: ", sklearn.__version__)
```

```python
np.random.seed(42)
X = np.random.rand(100, 5)

mean_all = X.mean()
std_all = X.std()
min_all = X.min()
max_all = X.max()

print(f"1. mean(среднее): {mean_all:.3f}")
print(f"2. std(стандартное отклонение): {std_all:.3f}")
print(f"3. min(минимальное): {min_all:.3f}")
print(f"4. max(максимальное): {max_all:.3f}")
print()

mean_cols = X.mean(axis=0)
std_cols = X.std(axis=0)
min_cols = X.min(axis=0)
max_cols = X.max(axis=0)

print(f"1. mean(среднее): {mean_cols}")
print(f"2. std(стандартное отклонение): {std_cols}")
print(f"3. min(минимальное): {min_cols}")
print(f"4. max(максимальное): {max_cols}")
print()

w = np.random.rand(5)
y = X @ w
print(f"vector y: {y.shape}")
```

```python
df = pandas.DataFrame(X, columns = ['price', 'quantity', 'discount', 'rating', 'reviews'])
df['target'] = y

print('head(первые 5 строк)')
print(df.head())
print()
print('shape(форма таблицы)')
print(df.shape)
print()
print('describe(описательная статистика)')
print(df.describe())
print()
print('isnull(проверка на пропуски)')
print(df.isnull().sum())
print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.figure(figsize=(8,5))
plt.hist(df['target'], bins=20, color = 'skyblue', edgecolor='black')
plt.title('target')
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()

sns.set(style = 'whitegrid')
corr_matrix = df[['price', 'quantity', 'discount', 'rating', 'reviews', 'target']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('heat map')
plt.show()
```
